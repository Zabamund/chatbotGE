## Data extraction - Liste de vote
#### Entities for LISTS in GC and CE

```python
# import libraries
import pandas as pd
```

```python
# import files
df_raw_CE = pd.read_json('./DATA_format_JSON/CE/datas/2351/CE_infosListes.json')
df_raw_GC = pd.read_json('./DATA_format_JSON/GC/datas/2353/GC_infosListes.json')
```

```python
df_raw_CE.head()
```

```python
df_raw_GC.head()
```

```python
# create Entity dfs (entity,value,synonyms)
# create duplicate DF by dropping one column in original
df_CE = df_raw_CE.drop(['noDepot'], axis=1)
# rename colmun
df_CE.rename(index=str, columns={"nom": "value"}, inplace=True)
# add entity column
df_CE['entity'] = 'Listes_CE'
# create synonyms colume
df_CE['synonyms'] = df_CE['nomComplet'].str.lower() + ":" + df_CE['sigle'].str.lower() + ":" + df_CE['id'].astype(str)
df_CE.head()
```

```python
# create Entity dfs (entity,value,synonyms)
# create duplicate DF by dropping one column in original
df_GC = df_raw_GC.drop(['noDepot'], axis=1)
# rename colmun
df_GC.rename(index=str, columns={"nom": "value"}, inplace=True)
# add entity column
df_GC['entity'] = 'Listes_GC'
# create synonyms colume
df_GC['synonyms'] = df_GC['nomComplet'].str.lower() + ":" + df_GC['sigle'].str.lower() + ":" + df_GC['id'].astype(str)
df_GC.head()
```

```python
# remove spaces
df_CE['synonyms'].replace(regex=True,inplace=True,to_replace=' ',value=':')
# remove accents from synonyms
df_CE['synonyms'].replace(regex=True,inplace=True,to_replace='é',value='e')
df_CE['synonyms'].replace(regex=True,inplace=True,to_replace='è',value='e')
df_CE['synonyms'].replace(regex=True,inplace=True,to_replace='ê',value='e')
df_CE['synonyms'].replace(regex=True,inplace=True,to_replace='à',value='a')
df_CE['synonyms'].replace(regex=True,inplace=True,to_replace='ï',value='i')
df_CE['synonyms'].replace(regex=True,inplace=True,to_replace='î',value='i')
df_CE['synonyms'].replace(regex=True,inplace=True,to_replace='ü',value='u')
df_CE['synonyms'].replace(regex=True,inplace=True,to_replace='ö',value='o')
# remove spec chars from synonyms
df_CE['synonyms'].replace(regex=True,inplace=True,to_replace="'",value='_')
df_CE['synonyms'].replace(regex=True,inplace=True,to_replace="-",value=':')
df_CE['synonyms'].replace(regex=True,inplace=True,to_replace=" ",value='_')
df_CE['synonyms'].replace(regex=True,inplace=True,to_replace="/",value='')
df_CE['synonyms'].replace(regex=True,inplace=True,to_replace="!",value='')
# remove duplicate ':'
df_CE['synonyms'].replace(regex=True,inplace=True,to_replace=':::',value=':')
df_CE['synonyms'].replace(regex=True,inplace=True,to_replace='::',value=':')

df_CE.head()
```

```python
# remove spaces
df_GC['synonyms'].replace(regex=True,inplace=True,to_replace=' ',value=':')
# remove accents from synonyms
df_GC['synonyms'].replace(regex=True,inplace=True,to_replace='é',value='e')
df_GC['synonyms'].replace(regex=True,inplace=True,to_replace='è',value='e')
df_GC['synonyms'].replace(regex=True,inplace=True,to_replace='ê',value='e')
df_GC['synonyms'].replace(regex=True,inplace=True,to_replace='à',value='a')
df_GC['synonyms'].replace(regex=True,inplace=True,to_replace='ï',value='i')
df_GC['synonyms'].replace(regex=True,inplace=True,to_replace='î',value='i')
df_GC['synonyms'].replace(regex=True,inplace=True,to_replace='ü',value='u')
df_GC['synonyms'].replace(regex=True,inplace=True,to_replace='ö',value='o')
# remove spec chars from synonyms
df_GC['synonyms'].replace(regex=True,inplace=True,to_replace="'",value='_')
df_GC['synonyms'].replace(regex=True,inplace=True,to_replace="-",value=':')
df_GC['synonyms'].replace(regex=True,inplace=True,to_replace=" ",value='_')
df_GC['synonyms'].replace(regex=True,inplace=True,to_replace="/",value='')
df_GC['synonyms'].replace(regex=True,inplace=True,to_replace="!",value='')
df_GC['synonyms'].replace(regex=True,inplace=True,to_replace="plr.",value='plr')
# remove duplicate ':'
df_GC['synonyms'].replace(regex=True,inplace=True,to_replace=':::',value=':')
df_GC['synonyms'].replace(regex=True,inplace=True,to_replace='::',value=':')

df_GC.head()
```

```python
# drop unused columns
df_CE.drop(['id', 'nomComplet', 'sigle'], axis=1, inplace=True)
df_CE.head()
```

```python
# drop unused columns
df_GC.drop(['id', 'nomComplet', 'sigle'], axis=1, inplace=True)
df_GC.head()
```
