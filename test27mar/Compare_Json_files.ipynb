```python
# reorder columns
cols = [('First', 'fonction'), ('Second', 'fonction'), ('First', 'liste'), ('Second', 'liste')
        , ('First', 'nom'), ('Second', 'nom'),('First', 'photoUrl'),('Second', 'photoUrl')
        , ('First', 'sexe'),('Second', 'sexe')]
df_reorder = df_all[cols]
df_reorder.head()
```

```python
def highlight_diff(data, color='yellow'):
    attr = 'background-color: {}'.format(color)
    other = data.xs('First', axis='columns', level=-1)
    return pd.DataFrame(np.where(data.ne(other, level=0), attr, ''),
                        index=data.index, columns=data.columns)

df_reorder.style.apply(highlight_diff, axis=None)
```
