## Data extraction - Liste de vote
#### Entities for LISTS in GC and CE

```python
# import libraries
import pandas as pd
```

```python
# import files
df_raw_CE = pd.read_json('./DATA_format_JSON/CE/datas/2351/CE_infosListes.json')
df_raw_GC = pd.read_json('./DATA_format_JSON/GC/datas/2353/GC_infosListes.json')
```

```python
df_raw_CE.head(1)
```

```python
df_raw_GC.head(1)
```

```python
# append two dfs - reindex
df = df_raw_CE.append(df_raw_GC, ignore_index=True, verify_integrity=True)
# drop noDepot
df.drop(labels=['noDepot', 'id'], axis=1, inplace=True, errors='raise')
# remove duplicate rows
df.drop_duplicates(subset=None, keep='first', inplace=True)
```

```python
df.head(30)
```

```python
# add value,entity,synonyms
df['value'] = df['sigle']
df['entity'] = 'ListeElectorale' 
df['synonyms'] = df['nom']
```

```python
# title case
df['synonyms'] = df['synonyms'].str.title()
# reset title for 3 letters
df['synonyms'][3] = df['value'][3].upper()
df['synonyms'][4] = 'PLR - PDC - Entente'
df['synonyms'][7] = df['value'][7].upper()
df['synonyms'][11] = 'PBD Genève'
```

```python
df.head(26)
```

```python
# drop unused columns
df.drop(['nom', 'nomComplet', 'sigle'], axis=1, inplace=True)
df.head()
```

```python
# export *.csv
df.to_csv('./electionBot_sBox/GBarthelet_chatbot/entities/liste-Entities.csv',index=False)
```
