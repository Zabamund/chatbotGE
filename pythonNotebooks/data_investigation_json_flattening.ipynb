```python
# import libraries
import pandas as pd
import json
from pandas.io.json import json_normalize
```

### Conseil d'Etat

```python
# import files
df_CE_infosCandidat = pd.read_json('./DATA_format_JSON/CE/datas/2351/CE_infosCandidat.json')
df_CE_infosCandidat.head()
```

```python
# flatten a nested part of nested json
df_CE_infosCandidat_liste = json_normalize(df_CE_infosCandidat['liste'])
print(df_CE_infosCandidat_liste.to_dict(orient='records')[0])
```

```python
# import files
df_CE_infosListes = pd.read_json('./DATA_format_JSON/CE/datas/2351/CE_infosListes.json')
df_CE_infosListes.head()
```

```python
# import nested json files
# flatten nested json
with open('./DATA_format_JSON/CE/datas/2351/CE_Resultats_Candidat_20001.json') as f:
    d = json.load(f)
    
df_CE_Resultats_Candidat_20001 = json_normalize(d)
df_CE_Resultats_Candidat_20001.head()
```

```python
# flatten a nested part of nested json
df_CE_Resultats_Candidat_20001_resultatsCandidatParCommune = json_normalize(d['resultatsCandidatParCommune'])
df_CE_Resultats_Candidat_20001_resultatsCandidatParCommune.head()
```

```python
# check aggregation
df_CE_Resultats_Candidat_20001_resultatsCandidatParCommune['suffrages'].sum()
```

```python
# import nested json files
with open('./DATA_format_JSON/CE/datas/2351/CE_Resultats_Canton.json') as f:
    d = json.load(f)
    
df_CE_Resultats_Canton = json_normalize(d)
df_CE_Resultats_Canton.head()
```

```python
# flatten a nested part of nested json
df_CE_Resultats_Canton_resultatsNominatifs = json_normalize(d['resultatsNominatifs'])
df_CE_Resultats_Canton_resultatsNominatifs.head()
```

```python
# import files
df_CE_Resultats_CartesSuffragesCandidat = pd.read_json('./DATA_format_JSON/CE/datas/2351/CE_Resultats_CartesSuffragesCandidat.json')
df_CE_Resultats_CartesSuffragesCandidat.head()
```

```python
# flatten a nested part of nested json
df_CE_Resultats_CartesSuffragesCandidat_candidat = json_normalize(df_CE_Resultats_CartesSuffragesCandidat['candidat'])
df_CE_Resultats_CartesSuffragesCandidat_candidat.head()
```

```python
# flatten a nested part of nested json
df_CE_Resultats_CartesSuffragesCandidat_carteSuffrageCandidat = df_CE_Resultats_CartesSuffragesCandidat['carteSuffrageCandidat']
df_CE_Resultats_CartesSuffragesCandidat_carteSuffrageCandidat[0]
```

```python
# import files
with open('./DATA_format_JSON/CE/datas/2351/CE_Resultats_Commune_0100.json') as f:
    d = json.load(f)
    
df_CE_Resultats_Commune_0100 = json_normalize(d)
df_CE_Resultats_Commune_0100.head()
```

```python
# flatten a nested part of nested json
df_CE_Resultats_Commune_0100_resultatsNominatifs = json_normalize(d['resultatsNominatifs'])
df_CE_Resultats_Commune_0100_resultatsNominatifs.head()
```

```python
# import files
df_CE_Resultats_InfoAvancement = pd.read_json('./DATA_format_JSON/CE/datas/2351/CE_Resultats_InfoAvancement.json')
df_CE_Resultats_InfoAvancement.head()
```

```python
# unstack a nested part of nested json
df_CE_Resultats_Commune_0100_resultatsNominatifs_Avancement = json_normalize(df_CE_Resultats_InfoAvancement['Avancement']).unstack()
df_CE_Resultats_Commune_0100_resultatsNominatifs_Avancement.head()
```

```python
# unstack a nested part of nested json
df_CE_Resultats_Commune_0100_resultatsNominatifs_Local = json_normalize(df_CE_Resultats_InfoAvancement['Local']).unstack()
df_CE_Resultats_Commune_0100_resultatsNominatifs_Local.head()
```
