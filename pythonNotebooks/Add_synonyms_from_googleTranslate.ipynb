## Adding google Translate synonyms to entities
This is required because of translation

```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# import files
df_candidates = pd.read_csv('./electionBot_sBox/archive/candidates-Entities_v3_edt.csv')
df_commune = pd.read_csv('./electionBot_sBox/archive/communes-Entities.csv')
df_liste_CE = pd.read_csv('./electionBot_sBox/archive/liste-CE-Entities.csv')
df_liste_GC = pd.read_csv('./electionBot_sBox/archive/liste-GC-Entities.csv')
df_locals = pd.read_csv('./electionBot_sBox/archive/locals-Entities.csv')
```

```python
# all files toLower for GoogleTranslate
df_candidates['synonyms'] = df_candidates['synonyms'].str.lower()
df_commune['synonyms'] = df_commune['synonyms'].str.lower()
df_liste_CE['synonyms'] = df_liste_CE['synonyms'].str.lower()
df_liste_GC['synonyms'] = df_liste_GC['synonyms'].str.lower()
df_locals['synonyms'] = df_locals['synonyms'].str.lower()
```

```python
# export csv files for use in GoogleTranslate
df_candidates.to_csv('./electionBot_sBox/entities/candidates-Entities_v5.csv',index=False)
df_commune.to_csv('./electionBot_sBox/entities/communes-Entities_v3.csv',index=False)
df_liste_CE.to_csv('./electionBot_sBox/entities/liste-CE-Entities_v3.csv',index=False)
df_liste_GC.to_csv('./electionBot_sBox/entities/liste-GC-Entities_v3.csv',index=False)
df_locals.to_csv('./electionBot_sBox/entities/locals-Entities_v3.csv',index=False)
```

```python
# add synonyms to df_candidates
df_candidates.head()
```

```python
# import googleTranslates
df_candidatesTranslates = pd.read_csv('./electionBot_sBox/googleTranslateEntities/candidates_translates.csv', delimiter=':', header=None)
df_candidatesTranslates.head()
```

```python
# add synonyms to df_candidates
df_candidates['synonyms'] = (df_candidates['synonyms']
                           + ':' + df_candidatesTranslates[0].astype(str) 
                           + ':' + df_candidatesTranslates[1].astype(str)
                           + ':' + df_candidatesTranslates[2].astype(str) 
                           + ':' + df_candidatesTranslates[3].astype(str)
                           + ':' + df_candidatesTranslates[4].astype(str)
                           + ':' + df_candidatesTranslates[5].astype(str)
                           + ':' + df_candidatesTranslates[6].astype(str))
df_candidates['synonyms'].replace(regex=True,inplace=True,to_replace='nan',value='')
df_candidates['synonyms'].replace(regex=True,inplace=True,to_replace=' ',value=':')
df_candidates['synonyms'].replace(regex=True,inplace=True,to_replace="'",value=':')
df_candidates['synonyms'].replace(regex=True,inplace=True,to_replace=':::',value=':')
df_candidates['synonyms'].replace(regex=True,inplace=True,to_replace='::',value=':')
df_candidates['synonyms'].replace(regex=True,inplace=True,to_replace='.0',value='')
df_candidates['synonyms'].replace(regex=True,inplace=True,to_replace='::',value=':')
df_candidates['synonyms'] = df_candidates['synonyms'].str.rstrip('\:')
```
